```python
#Retrieve the Item ID, Item Name, and Item Price columns
item_name_id_price = purchase_data[["Item ID","Item Name","Price"]]
item_name_id_price
```

```python
#Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
group_id_name = item_name_id_price.groupby(["Item ID","Item Name"])
```

```python
#total purchase value
total_purchase_value = group_id_name.sum()
final_summary_df = total_purchase_value.rename(columns = {"Price":"Total Purchase Value"})
final_summary_df
```

```python
#purchase count
x = purchase_data.groupby(["Item ID"])["Item Name"].value_counts()
print(x)
purchase_count_df2 = pd.DataFrame(x).rename(columns={"Item Name":"Purchase Count"})
purchase_count_df2 
```

```python
final_summary_df = final_summary_df.merge(purchase_count_df2,left_index = True, right_index = True)
final_summary_df
```

```python
#Individual Item Price
indi_item_price = round(final_summary_df["Total Purchase Value"]/final_summary_df["Purchase Count"],2)
final_summary_df["Item Price"] = indi_item_price
```
